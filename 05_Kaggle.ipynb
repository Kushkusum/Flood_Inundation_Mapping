```python
import rasterio, json, os
from tqdm import tqdm

S1_TRAIN = '/content/drive/MyDrive/FloodProject/data/sen1floods11/train/S1Weak'
S1_VAL   = '/content/drive/MyDrive/FloodProject/data/sen1floods11/val/S1Hand'

bounds_dict = {}

for s1_dir, split in [(S1_TRAIN, 'train'), (S1_VAL, 'val')]:
    for fname in tqdm(os.listdir(s1_dir), desc=split):
        if not fname.endswith('.tif'):
            continue
        chip_id = fname.replace('_S1Weak.tif','').replace('_S1Hand.tif','')
        with rasterio.open(os.path.join(s1_dir, fname)) as src:
            b = src.bounds
            bounds_dict[chip_id] = {
                'left'  : b.left,
                'bottom': b.bottom,
                'right' : b.right,
                'top'   : b.top,
                'split' : split
            }

out_path = '/content/drive/MyDrive/FloodProject/data/sen1floods11/metadata/chip_bounds.json'
with open(out_path, 'w') as f:
    json.dump(bounds_dict, f)

print(f"Saved bounds for {len(bounds_dict)} chips")
```
