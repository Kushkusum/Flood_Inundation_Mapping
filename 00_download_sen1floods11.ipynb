***Download Train S2Weak***


```python
import subprocess, os, glob

BASE = "/content/drive/MyDrive/FloodProject/data/sen1floods11"
DEST = f"{BASE}/train/S2Weak"
os.makedirs(DEST, exist_ok=True)

# Verify destination
assert os.path.exists(DEST), f"ERROR: {DEST} does not exist"
print(f"Destination verified: {DEST}\n")

COUNTRIES = ["India", "Pakistan", "Sri-Lanka", "Mekong", "Bolivia", "Colombia"]

for country in COUNTRIES:
    print(f"Downloading S2Weak: {country}...")
    src = f"gs://sen1floods11/v1.1/data/flood_events/WeaklyLabeled/S2Weak/{country}_*_S2Weak.tif"

    result = subprocess.run(
        ["gsutil", "-m", "cp", src, f"{DEST}/"],
        capture_output=False,   # let output stream live to notebook
        text=True
    )

    count = len(glob.glob(f"{DEST}/{country}_*_S2Weak.tif"))
    status = "✓" if count > 0 else "✗ FAILED"
    print(f"  -> {count} chips  {status}\n")

print("=== Train S2Weak final counts ===")
total = 0
for country in COUNTRIES:
    n = len(glob.glob(f"{DEST}/{country}_*_S2Weak.tif"))
    total += n
    print(f"  {country:<12}: {n} chips")
print(f"  {'TOTAL':<12}: {total} chips")
print(f"\nExpected: ~3017  |  Got: {total}  {'✓' if total >= 3000 else '✗'}")
```

***Full Verification***

```python
import os, glob, rasterio

BASE = "/content/drive/MyDrive/FloodProject/data/sen1floods11"

train_countries = ["India", "Pakistan", "Sri-Lanka", "Mekong", "Bolivia", "Colombia"]
val_countries   = ["India", "Pakistan", "Sri-Lanka", "Mekong"]

# ── 1. Training set counts ────────────────────────────────────
print("=" * 55)
print("TRAINING SET (WeaklyLabeled)")
print("=" * 55)
total_train = 0
for country in train_countries:
    s1 = len(glob.glob(f"{BASE}/train/S1Weak/{country}_*_S1Weak.tif"))
    s2 = len(glob.glob(f"{BASE}/train/S2Weak/{country}_*_S2Weak.tif"))
    lb = len(glob.glob(f"{BASE}/train/S2IndexLabelWeak/{country}_*_S2IndexLabelWeak.tif"))
    ok = "✓" if s1 == s2 == lb and s1 > 0 else "✗ MISMATCH"
    total_train += s1
    print(f"  {country:<12}: S1={s1}  S2={s2}  Label={lb}  {ok}")
print(f"  {'TOTAL':<12}: {total_train} chips")

# ── 2. Validation set counts ──────────────────────────────────
print()
print("=" * 55)
print("VALIDATION SET (HandLabeled)")
print("=" * 55)
total_val = 0
for country in val_countries:
    s1 = len(glob.glob(f"{BASE}/val/S1Hand/{country}_*_S1Hand.tif"))
    s2 = len(glob.glob(f"{BASE}/val/S2Hand/{country}_*_S2Hand.tif"))
    lb = len(glob.glob(f"{BASE}/val/LabelHand/{country}_*_LabelHand.tif"))
    ok = "✓" if s1 == s2 == lb and s1 > 0 else "✗ MISMATCH"
    total_val += s1
    print(f"  {country:<12}: S1={s1}  S2={s2}  Label={lb}  {ok}")
print(f"  {'TOTAL':<12}: {total_val} chips")

# ── 3. Band count spot check ──────────────────────────────────
print()
print("=" * 55)
print("BAND COUNT SPOT CHECK")
print("=" * 55)
checks = [
    (f"{BASE}/train/S1Weak",           "India", 2,  "VV+VH, float32"),
    (f"{BASE}/train/S2Weak",           "India", 13, "B1-B12+QA, uint16"),
    (f"{BASE}/train/S2IndexLabelWeak", "India", 1,  "label 0/1, int16"),
    (f"{BASE}/val/S1Hand",             "India", 2,  "VV+VH, float32"),
    (f"{BASE}/val/S2Hand",             "India", 13, "B1-B12+QA, int16"),
    (f"{BASE}/val/LabelHand",          "India", 1,  "-1/0/1, int16"),
]
for folder, country, exp_bands, desc in checks:
    files = sorted(glob.glob(f"{folder}/{country}_*.tif"))
    if not files:
        print(f"  {os.path.basename(folder):<25}: NO FILES ✗")
        continue
    with rasterio.open(files[0]) as src:
        ok = "✓" if src.count == exp_bands else f"✗ got {src.count}"
        print(f"  {os.path.basename(folder):<25}: {src.count} bands  "
              f"dtype={src.dtypes[0]:<8}  {src.height}x{src.width}  {ok}  ({desc})")

# ── 4. Storage summary ────────────────────────────────────────
print()
print("=" * 55)
print("STORAGE SUMMARY")
print("=" * 55)
folders = {
    "train/S1Weak":           f"{BASE}/train/S1Weak",
    "train/S2Weak":           f"{BASE}/train/S2Weak",
    "train/S2IndexLabelWeak": f"{BASE}/train/S2IndexLabelWeak",
    "val/S1Hand":             f"{BASE}/val/S1Hand",
    "val/S2Hand":             f"{BASE}/val/S2Hand",
    "val/LabelHand":          f"{BASE}/val/LabelHand",
}
total_gb = 0
for name, folder in folders.items():
    tifs = glob.glob(f"{folder}/*.tif")
    gb   = sum(os.path.getsize(f) for f in tifs) / 1e9
    total_gb += gb
    print(f"  {name:<25}: {len(tifs):>5} files  {gb:>6.2f} GB")
print(f"  {'TOTAL':<25}: {total_gb:>6.2f} GB")
```
